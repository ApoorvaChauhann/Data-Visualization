```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data (3D scatter plot)
x = np.random.rand(100) * 10  # Random values between 0 and 10
y = np.random.rand(100) * 10
z = np.sin(x) * np.cos(y)  # Create a 3D pattern

# Configure the 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
ax.scatter(x, y, z, c='red', marker='o')  # Red spheres

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')

# Set viewing angles (adjust as needed)
ax.view_init(elev=15, azim=60)  # Elevation and azimuth angles

# Display the plot
plt.show()


```
